#### Importing libraries
Write the code to import Tic-Tac-Toe class from the environment file

```python
from TCGame_Env import TicTacToe ## - import your class from environment file
import collections
import numpy as np
import random
import pickle
import time
from matplotlib import pyplot as plt
```

```python
# Function to convert state array into a string to store it as keys in the dictionary
# states in Q-dictionary will be of form: x-4-5-3-8-x-x-x-x
#   x | 4 | 5
#   ----------
#   3 | 8 | x
#   ----------
#   x | x | x

def Q_state(state):
    return ('-'.join(str(e) for e in state)).replace('nan','x')
```

```python
# Defining a function which will return valid (all possible actions) actions corresponding to a state
# Important to avoid errors during deployment.

def valid_actions(state):
    valid_Actions = []
    valid_Actions = [i for i in env.action_space(state)[0]] ###### -------please call your environment as env
    return valid_Actions
```

```python
# Defining a function which will add new Q-values to the Q-dictionary. 
def add_to_dict(state):
    state1 = Q_state(state)
    
    valid_act = valid_actions(state)
    if state1 not in Q_dict.keys():
        for action in valid_act:
            Q_dict[state1][action]=0
```

#### Epsilon-greedy strategy - Write your code here

(you can build your epsilon-decay function similar to the one given at the end of the notebook)

```python
# Defining epsilon-greedy policy. You can choose any function epsilon-decay strategy
def epsilon_greedy_func(cur_state, time):
    epsilon = min_epsilon + (max_epsilon - min_epsilon) * np.exp(-decay_rate*time)
    z = np.random.random()
    
    if z > epsilon:  #Exploitation : this gets the action corresponding to max q-value of current state
        state1 = Q_state(cur_state)
        new_action = max(Q_dict[state1],key=Q_dict[state1].get) 
    else: #Exploration :randomly choosing an action
        possible_actions = [action for action in env.action_space(cur_state)[0]]        
        new_action = possible_actions[np.random.choice(range(len(possible_actions)))]
    
    return new_action
```

#### Tracking the state-action pairs for checking convergence - write your code here

```python
# Initialise Q_dictionary as 'Q_dict' and States_tracked as 'States_track' (for convergence)
Q_dict = collections.defaultdict(dict)
States_track =collections.defaultdict(dict)
```

```python
# Initialise states to be tracked
state1 = 'x-x-x-x-x-x-x-x-x'
action1 = (0,1)
state2 = 'x-x-x-1-x-x-2-x-x'
action2 = (1,9)
state3 = 'x-1-x-x-4-x-x-x-x'
action3 = (7,5)
state4 = 'x-x-x-x-x-x-x-x-x'
action4 = (5,7)

def initialise_tracking_states():
    sample_q_values = [(state1,action1),(state2,action2),(state3,action3),(state4,action4)]  
    for q_values in sample_q_values:
        state = q_values[0]
        action = q_values[1]
        States_track[state][action] = []    #this is an array which will have appended values of that state-action pair for every 200th episode         
```

```python
#Defining a function to save the Q-dictionary as a pickle file

def save_obj(obj, name ):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
```

```python
def save_tracking_states():
    for state in States_track.keys():
        for action in States_track[state].keys():
            if state in Q_dict and action in Q_dict[state]:
                States_track[state][action].append(Q_dict[state][action])
```

#### Define hyperparameters  ---write your code here

```python
EPISODES = 500000
# learning rate
LR = 0.01
# Discount factor
GAMMA = 0.91               
max_epsilon = 1.0
min_epsilon = 0.001
decay_rate = 0.000001
# every these 200 episodes, the 4 Q-values will be stored/appended (convergence graphs)
threshold = 200
#every 300 episodes, the Q-dict will be updated
policy_threshold = 300   
```

### Q-update loop ---write your code here

```python
start_time = time.time()

for episode in range(EPISODES):
    ##### Start writing your code from the next line
    
    #Initialize environment and state
    env = TicTacToe()
    curr_state = env.state    
    isTerminated = False  
    add_to_dict(curr_state)

    while isTerminated != True:  
        curr_state_temp = Q_state(curr_state)
        curr_action = epsilon_greedy_func(curr_state, episode)   # applying epislon method
        next_state, reward, isTerminated = env.step(curr_state, curr_action) # getting rewar
        next_state_temp = Q_state(next_state)
        add_to_dict(next_state)

        # Updating Q-table
        if isTerminated != True:
            max_next = max(Q_dict[next_state_temp],key=Q_dict[next_state_temp].get)  
            #action corresponding to max q-value for the next state
            Q_dict[curr_state_temp][curr_action] += LR * ((reward + (GAMMA*(Q_dict[next_state_temp][max_next])))
                                                      - Q_dict[curr_state_temp][curr_action] ) 
        else:
            Q_dict[curr_state_temp][curr_action] += LR * ((reward - Q_dict[curr_state_temp][curr_action]))

        #Change State
        curr_state = next_state
        
    #TRACKING Q-VALUES
    if (episode == threshold-1):        #at the 199th episode
        initialise_tracking_states()
        
    if ((episode+1)%threshold)==0:#every 200th episodes, save states
        #tracking states 
        save_tracking_states()
        save_obj(States_track,'States_tracked')
    
    #save policy
    if ((episode+1)% policy_threshold ) == 0:  #every 300th episodes, the Q-dict will be saved
        save_obj(Q_dict,'Policy')      
    
    
elapsed_time = time.time() - start_time
save_obj(States_track,'States_tracked')   
save_obj(Q_dict,'Policy')
```
